1. Seleksi Unvariate

```python
# import library
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# memuat data
data = pd.read_csv("stock.csv")
data.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

#baca file data
df = pd.read_csv('stock.csv')

#inisialisasi objek labelEncoder
label_encoder = LabelEncoder()

#Lakukan encoding untuk setiap kolom non-numerik
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = label_encoder.fit_transform(df[col])

#Simpan data dalam file baru
df.to_csv('stock_encoded.csv',index=False)
```

```python
# memuat data
data = pd.read_csv("stock_encoded.csv")
data.head()
```

```python
# memilih data yang dibutuhkan
X = data.iloc[:,0:20] 
y = data.iloc[:,-1] 
```

```python
# menerapkan SelectKBest untuk melakukan ekstraksi
bestfeatures = SelectKBest(score_func=chi2, k=8)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
# menggabungkan 2 dataframe
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Type','Score'] 
print(featureScores.nlargest(5,'Score')) 
```

2. Feature Importance

```python
# import library
import pandas as pd
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
```

```python
# memuat data
data = pd.read_csv("stock_encoded.csv")
X = data.iloc[:,0:20] 
y = data.iloc[:,-1] 
```

```python
# melakukan ExtraTreesClassifier untuk mengekstraksi fitur
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) 
```

```python
# melakukan plot dari feature importances
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```
